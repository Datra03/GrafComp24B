# MODELO LOTKA-VOLTERRA

## Es un modelo el cual supone a dos especies en interacción (presa y despredador) las cuales describen como van cambiando sus poblaciones a lo largo del tiempo

### Modelo en python 

a = Aumento natural
b = Tasa de depredación 
c = Muerte en ausencia de presas 
e = Aumento causado por alimentación disponible 

El siguente codigo es un ejemplo de como funciona el modelo lotka-volterra donde las presas y los depredsadore interactuan en un mismo medio y se tiene una armonia donde ambas especies estan en control, ambas crecen, se reproducen y son cazadas o fallecen

```python
%matplotlib inline
import matplotlib.pyplot as plt
from random import *
from numpy import *
import sys
```

```python
a = 0.7; b = 0.5; c = 0.3; e = 0.2
```

```python
dt = 0.001; max_time = 100
```

```python
t = 0; x = 0.1; y = 0.5
```

```python
t_list = []; x_list = []; y_list = []
```

```python
t_list.append(t); x_list.append(x); y_list.append(y) 
```

```python
while t < max_time:
    t = t + dt
    x = x + (a*x - b*x*y)*dt
    y = y + (-c*y + e*x*y)*dt
    
    t_list.append(t)
    x_list.append(x)
    y_list.append(y)
```

```python
p = plt.plot(t_list, x_list, 'r', t_list, y_list, 'g', linewidth = 2)
```

### CASO 1

En este caso, si aumentamos el parametro b (tasa de depredacion) lo que observaremos es un aumento exponencial de las presas, los depredadores crecen y cazan, pero son rebasados por las presas, mostrando una sobrepoblacion de esta especie
